```python
class Expence:
    def __init__(self, info):
        self.info =info
    def add_expence(self, category, name, date, amt):
        global id
        self.info[id] = {'category':category,'name':name,'date':date,'amt':amt}
        print("Expence add Successfully.")
        
        int_id = int(id)
        int_id+=1
        id = str(int_id)

    def view_full (self):
        print(self.info)
    
    def veiw_date(self, date_in):
        for i in self.info.keys():
            if self.info[i]['date'] == date_in:
                print(self.info[i])
            else :
                print("Expence not found")
                
    def view_date_range (self, fromDate, toDate):
        for i in self.info.keys():
            date_format='%Y-%m-%d'
            in_fromDate = datetime.strptime(fromDate, date_format).date()
            in_toDate = datetime.strptime(toDate, date_format).date()
            expense_date = datetime.strptime(self.info[i]['date'], date_format).date()
            if in_fromDate <= expense_date <= in_toDate:
                print(self.info[i])
            else:
                continue
        else:
            print('No expenses made on this range')
    def view_amt(self, amount):
        for i in self.info.keys():
            if int(amount) == int(self.info[i]['amt']):
                print(self.info[i])
            else:
                print("Expence not found")
    def view_category(self, category):
        for i in self.info.keys():
            if self.info[i]['category'] == category:
                print(self.info[i])
            else:
                print("Expence not found")
    
```
