```python
#import module json 
import json
```

```python
#load the data from bank.json file 
with open ("bank.json", 'r') as f:
    data = json.load(f)
```

```python
#split the data into two part
ci = data['customer_information']
tr = data['Transaction']
print(ci)
print(tr)
```

```python
class Bank:#class bank 
    def __init__(self, ci, tr):#construct ci and tr from the gloabal variable
        self.ci = ci
        self.tr = tr
    
    def user_add(self,ac_no,name,balance):
        self.ci[ac_no]={'name':name,'balance':balance}
        print("THE DETAILS HAVE BEEN UPDATED: ")
        print(ci)
    
    def delete(self,ac_no):
        close_ac=self.ci.pop(ac_no)
        print("the closed account is :",close_ac)
        print("THE UPDATED DETAILS ARE: ")
        print(ci)
    
    def trans(self, fromacc, toacc, amount):#transfer fuction 
        if str(fromacc) not in self.ci.keys() :
            print(fromacc," not found....")
        if str(toacc) not in self.ci.keys():
            print(toacc," not found....")
        if str(fromacc) in self.ci.keys() and str(toacc) in self.ci.keys():
            if self.ci[str(fromacc)]['balance'] >= amount:#check sufficient balance 
                self.ci[str(fromacc)]['balance'] -= amount#decrease balance from sender
                self.ci[str(toacc)]['balance'] += amount#increase balance to reciver 
                #print the success proccess
                print(amount," is tranfer to ",toacc," from ",fromacc)
                print(self.ci[str(fromacc)]['name']," account balance is ",self.ci[str(fromacc)]['balance'])
                print(self.ci[str(toacc)]['name']," account balance is", self.ci[str(toacc)]['balance'])
            else:
                print('insufficient balance ...')
    
    def get_balance(self, customer_id):
        return self.ci[str(customer_id)]['balance']
    #Transaction part
    def TransactionUsing_TransacID(self,transac_id):
        if transac_id in self.tr:               #transac_id denotes -> transaction id
            transac_details=self.tr[transac_id]   ##transac_details denotes -> transaction id details
            print("transaction id : ",transac_id ," , info :", transac_details)
            
        else:
            print("not available as such transac id")
    def Transac_Check_accNO(self,accID):
        for tr_id in self.tr:               #tr_id denotes -> transaction id
            tr_details=self.tr[tr_id]   ##tr_details denotes -> transaction id details
            if tr_details['from'] == accID or tr_details['to'] == accID:
                print("transaction id : ",tr_id ,", info :", tr_details  )
                
        else: #check acc id/no exist in RbwebSkillup Bank or not
            for id in self.ci:    
                if int(id) == accID:
                    break    #accID exist in RbwebSkillup Bank 
                    
            else:
                print(accID,"-> this acc no. doesn't exist in RbwebSkillup Bank.")

    def deposit(self, to, amt):
        ci[str(to)]['balance'] += amt
        print(ci[str(to)]['name'], " account balance is ",ci[str(to)]['balance'])
    
    def New_trID(self):
        str_empty = []  #for string      #empty list
        dig_empty = []  #for digit       #empty list
        for ele in self.L_trID:
            if ele.isalpha():      
                str_empty.append(ele)   #append characters(strings) successively
            else:
                dig_empty.append(ele)   #append digits successively
                
        #string part
        str1=''         #empty string
        C_str_empty=[]  #empty list
        for i in str_empty:
            str1+=i 
        C_str_empty.append(str1)   #converted into single element list
        
        #digit part
        str2=''          #empty string
        C_dig_empty=[]   #empty list
        for j in dig_empty:
            str2+=j
        C_dig_empty.append(str2)


        z = int(C_dig_empty[0])     #z stores int value of C_dig_empty and proceed increament 1 then type casting to str 
        z+=1       #1 increament     #and store in l3 ->above comment's extension
        l3=[]      #empty list
        l3.append(str(z))
        C_str_empty.extend(l3)     #Two list add -> str_empty extends
        
        global N_trID              #makes global var to call anywhere
        N_trID  = ''.join(C_str_empty)
        
    #NewTransac add
    def NewTransac_Dt(self, to, amt):  #for deposit
        self.tr[N_trID] = {'from': "self branch", 'to': to , 'amt': amt}
    def NewTransac_Wt(self):  #for withdraw
        self.tr[N_trID] = {'from': "self_acc", 'to': "self", 'amt': 36520}  #in hand cash
    def NewTransac_Tr(self, fromacc, toacc, amount):  #for transfer
        self.tr[N_trID] = {'from': fromacc, 'to': toacc, 'amt': amount}

    def TransacCheckUsingAccName(self,accNAME):
        #print all the transaction for account holder name   #1234

        #customer information
        for n in self.ci: #n denotes ci id(as ACC ID/NO)    #here key(acc_no) is string note that
            ci_details=self.ci[n] #ci[n] denotes details as following id
            if ci_details['name'] == accNAME:
                print({accNAME},"'s acc no : ",{n})

                #Transaction 
                for p in self.tr: #p denotes trasac id
                    t_details=self.tr[p] #tr[p] denotes transaction element/details
                    if t_details['from'] == int(n) or t_details['to'] == int(n): #n(as ACC ID/NO) in ci -> string convert into int
                        print({accNAME},"'s Transaction id -> ",p," and details -> ",t_details)
                else:  
                    print({accNAME},"'s  no more transaction available")   
                    break
        else:
            print({accNAME} ,"-> acc holder doesn't exist in RbwebSkillup Bank") 
    
    def withdraw(self, ac_no, amount_to_withdraw): ## withdraw function
        if amount_to_withdraw < self.ci[str(ac_no)]['balance']: ## using if condition to check current balance after withdrawal
            self.ci[str(ac_no)]['balance'] -= amount_to_withdraw
            print({self.ci[str(ac_no)]['name']},"'s current or update balance is ",self.ci[str(ac_no)]['balance'])
        else:
            print('your balance is Insufficient !!')
        
   

```

```python
def main():
    user = Bank(ci, tr)
    print("Welcome to RbwebSkillup Bank")
    while(True):
        try:
            print("""What do you want 
            press 1 to add user
            press 2 to delete user
            press 3 to edit user details
            press 4 to check balance
            press 5 to deposit 
            press 6 to withdraw
            press 7 to transfer
            press 8 to check transaction
            press 9  to exit""")
            choise = int(input("\n Enter your choise: "))
            if choise == 1:
                try:
                    n=int(input("How many account you want to add: "))
                    for i in range (n):
                        ac_no=input("Enter account number: ")
                        name=input("enter your name: ")
                        balance=input("enter your balance:")
                        user.user_add(ac_no,name,balance)
                except ValueError as v:
                        print(v, "Enter valid input")
            elif choise == 2:
                try:
                    ac_no=input("Enter account no whose details you want to delete: ")
                    user.delete(ac_no)
                except ValueError as v:
                    print(v, "Enter valid input")
            elif choise == 3:
                pass
            elif choise == 4:
                try:
                    customer_id = int(input("Enter the customer_ID:"))
                    balance = user.get_balance(customer_id)
                    print(f"Customer {customer_id}'s balance: {balance}")
                except ValueError as v:
                    print(v, "Enter valid input ")
            elif choise == 5:
                try:
                    to = int(input("Enter accunt number: "))
                    amt = int(input("Enter ammount: "))
                    user.deposit(to, amt)
                except ValueError as v:
                    print(v, "Enter valid input: ")
            elif choise == 6:
                try:
                      ac_no = int(input("Enter account number: "))
                      amount_to_withdraw = int(input("Enter withdraw amount: "))
                      user.withdraw(ac_no,amount_to_withdraw)
                except ValueError as v:
                      print("Enter valid input", v)
            elif choise == 7:
                try:
                    fromacc = int(input("Enter accunt no oof sender: "))
                    toacc = int(input("Enter account no of reciver: "))
                    amount = int(input("Enter amount: "))
                    user.trans(fromacc, toacc, amount)
                except ValueError as v:
                    print(v, "Enter valid input")
            elif choise == 8:
                while True:
                    try:
                        print("""
                            Transactions check choice>>
                            press 1 to check transaction using transac id specificly
                            press 2 to check transactions using acc id/no
                            press 3 to check transactions using acc holder name

                            press 4 to exit""")
                        option_TR=int(input("\nEnter your choice : "))
                        match option_TR:
                            case 1:
                                try:    #check transaction of specific transac id
                                    transac_id=input("Enter transaction id : ")
                                    user.TransactionUsing_TransacID(transac_id)
                                except ValueError as t:
                                    print(t, "Enter valid transac id")
                            case 2:
                                try:    #check transactions using acc no/id
                                    accID=int(input("Enter acc id/no : "))
                                    user.Transac_Check_accNO(accID)
                                except ValueError as t1:
                                    print(t1, "Enter valid acc id .it should be no. ")
                            case 3:
                                try:    #check transactions of acc name
                                    accNAME=input("Enter acc name : ")
                                    user.TransacCheckUsingAccName(accNAME)
                                except ValueError as t2:
                                    print(t2, "Enter valid acc name")
                            case 4:
                                break
                            case _:
                                raise ValueError()
                    except ValueError as VE:
                        print(VE, "Enter valid input")
            elif choise == 9:
                break
            else:
                raise ValueError()
        except ValueError as v:
            print(v, "Enter valid input....")
```

```python
if __name__ == "__main__":
    main()
```
