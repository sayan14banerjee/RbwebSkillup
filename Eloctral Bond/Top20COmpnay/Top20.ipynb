```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('PurchaseData.csv')
```

```python
df.head(10)
```

```python
grouped = df.groupby('Purchaser Name')['Denomination'].sum()
```

```python
grouped
```

```python
top_20 = grouped.sort_values(ascending = False).head(20)
```

```python
top_20
```

```python
# Function to convert numbers to a shortened format
def shorten_number(number):
    if number >= 10**9:
        return f"{number // 10**9} bil"
    elif number >= 10**7:
        return f"{number // 10**7} cr"
    else:
        return number

top_20_plot = top_20.plot(kind='bar', figsize=(10, 6))

bars = top_20_plot.containers[0]

for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
             shorten_number(int(bar.get_height())), ha='center', va='bottom')

plt.title('Top 20 Purchasers by Total Denomination')
plt.xlabel('Purchaser Name')
plt.ylabel('Total Denomination')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
